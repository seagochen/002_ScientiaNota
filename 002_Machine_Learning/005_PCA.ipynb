```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
import numpy as np

# 
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
feature_names = iris.feature_names

# DataFrame
df = pd.DataFrame(X, columns=feature_names)
df['target'] = y

# 
sns.pairplot(df, vars=feature_names, hue='target', palette=['blue', 'red', 'green'], markers=['o', '^', 's'])
plt.show()

# 
cov_matrix = np.cov(X.T)

# 
sns.heatmap(cov_matrix, annot=True, cmap='Blues')
plt.show()

# 
iris = load_iris()
X = iris.data

# 
eigvals, eigvecs = np.linalg.eig(cov_matrix)

# 
print(':', eigvals)
print(':', eigvecs)


# PCA
pca = PCA(n_components=2)
X_new = pca.fit_transform(X)

# DataFrame
df_new = pd.DataFrame(X_new, columns=['PC1', 'PC2'])
df_new['target'] = y

# 
colors = ['blue', 'red', 'green']
markers = ['o', '^', 's']
fig, ax = plt.subplots()
for i, target_name in enumerate(target_names):
    ax.scatter(df_new[df_new['target'] == i]['PC1'],
               df_new[df_new['target'] == i]['PC2'],
               c=colors[i], marker=markers[i], label=target_name)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.legend()
plt.show()
```
